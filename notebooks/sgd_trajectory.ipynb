```python
import matplotlib.pyplot as plt
import numpy as np
from toy_sgd import get_trajectory_comparison
```

```python
n_train_samples = 128
concepts = [
    {"n_samples": 64, "column": 0, "seed": 0},
    {"n_samples": 64, "column": 1, "seed": 1},
]
trajectories = [
    {
        "batch_size": 4,
        "optimizer": "COCOB",
        "linestyle": "solid",
        "marker": "s",
        "color": "darkred",
        "marker_interval": 4,
        "lr": 10,
        "label": "COCOB",
    },
    {
        "batch_size": 4,
        "optimizer": "DoG",
        "linestyle": "dashed",
        "color": "darkorange",
        "marker": "o",
        "marker_interval": 4,
        "lr": 40,
        "label": "DoG",
    },
    {
        "batch_size": 4,
        "optimizer": "Adam",
        "linestyle": "dotted",
        "marker": "^",
        "marker_interval": 4,
        # "lr": 0.5,
        "label": "Adam",
    },
]

fig1, ax1 = plt.subplots(figsize=(3.3, 3))
fig2, ax2 = plt.subplots(figsize=(4.05, 3))
ax = np.array([ax1, ax2])
ax, contours = get_trajectory_comparison(
    concepts,
    trajectories,
    n_train_samples,
    contour_levels=20,
    axs=ax,
    # xlims=np.array([-3.5, 1.5]),
)
sm = plt.cm.ScalarMappable(cmap=contours[1].cmap, norm=contours[1].norm)

cbar = fig2.colorbar(sm, ax=ax[1])
cbar.ax.set_ylabel("Loss")
ax[0].legend(loc="lower right", frameon=True)
ax[0].set_xlabel(r"$\theta_1$", labelpad=4)
ax[1].set_xlabel(r"$\theta_1$", labelpad=3.5)
# axs[0].set_title("Concept 1", fontsize=12)
# axs[1].set_title("Concept 2", fontsize=12)
fig1.tight_layout()
fig2.tight_layout()

fig1.savefig("../pub/figures/sgd_trajectory_optims1.pdf", bbox_inches="tight")
fig2.savefig("../pub/figures/sgd_trajectory_optims2.pdf", bbox_inches="tight")
```
