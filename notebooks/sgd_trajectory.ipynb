```python
import matplotlib.pyplot as plt
import numpy as np
from toy_sgd import get_trajectory_comparison
```

```python
n_train_samples = 128
concepts = [
    {"n_samples": 64, "column": 0, "seed": 0, "cmap": "viridis"},
    {
        "n_samples": 64,
        "column": 1,
        "seed": 1,
        "cmap": "viridis",
        "origin_marker": "s",
        "origin_color": "#481467",
    },
]
trajectories = [
    {
        "batch_size": 4,
        "reset_at": None,
        "optimizer": "Adam",
        "color": "black",
        "linestyle": "dashed",
        "marker": "o",
        "marker_interval": 4,
        "label": "Adam",
        "arrow_size": 15,
    },
    {
        "batch_size": 4,
        "reset_at": 64,
        "optimizer": "Adam",
        "color": "black",
        "linestyle": "solid",
        "marker": "^",
        "marker_interval": 4,
        "arrow_size": 15,
        "label": "Adam w/ Drift-Reset",
    },
]
fig1, ax1 = plt.subplots(figsize=(5, 3))
fig2, ax2 = plt.subplots(figsize=(5, 3))
ax = np.array([ax1, ax2])
ax, contours = get_trajectory_comparison(
    concepts,
    trajectories,
    n_train_samples,
    axs=ax,
    # gridspec={"width_ratios": [1, 1.25]},
    contour_levels=15,
    # figsize=(4.5, 4),
)
sm = plt.cm.ScalarMappable(cmap=contours[1].cmap, norm=contours[1].norm)

cbar1 = plt.colorbar(sm, ax=ax[0])
cbar1.ax.set_ylabel("Loss")
cbar2 = plt.colorbar(sm, ax=ax[1])
cbar2.ax.set_ylabel("Loss")
ax[1].legend(loc="lower left")
ax[0].set_ylabel(r"$\theta_2$", labelpad=5)
ax[1].set_ylabel(r"$\theta_2$", labelpad=5)
fig1.tight_layout()
fig2.tight_layout()
fig1.savefig("../pub/figures/adam_trajectory_drift_reset1.pdf")
fig2.savefig("../pub/figures/adam_trajectory_drift_reset2.pdf")
```
